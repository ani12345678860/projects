```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
path='/content/drive/MyDrive/most_subscribed_youtube_channels.csv'
```

```python
df=pd.read_csv(path)
```

```python
print(type(df))
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['subscribers'] = df['subscribers'].str.replace(',', '').astype(int)
```

```python
df['video count'] = df['video count'].str.replace(',', '').astype(int)
```

```python
df['video views'] = df['video views'].str.replace(',', '').astype(int)
```

```python
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
sns.scatterplot(data=df, x='rank', y='subscribers')  #correlation between rank and subscribers
plt.title('rank vs subscribers')
plt.show()

correlation = df['rank'].corr(df['subscribers'])
print(f'Correlation between Rank and Subscribers: {correlation}')
```

```python
sns.barplot(data=df, x='category', y='subscribers',palette='viridis')    #distribution of subscribers across categories in which shows is the most liked by subscribers
plt.title('Distribution of Subscribers Across Categories')
plt.xticks(rotation=90)
plt.show()

category_stats = df.groupby('category')['subscribers'].agg(['mean', 'median']).sort_values(by='mean',ascending=False)
print(category_stats)
```

```python
df['No_of_years']=2024-df['started'] #finding age of movies to find growth rate
```

```python
df.head()
```

```python
df['growth_rate']=df['subscribers']/df['No_of_years'] #formula for finding growth_rate of a category
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='started', y='growth_rate')
plt.title('started vs growh_rate')
plt.show()

# Correlation analysis
correlation = df['started'].corr(df['growth_rate'])
print(f'Correlation between started year and growth_rate: {correlation}')
```

```python
sns.scatterplot(data=df, x='video count', y='subscribers')
plt.title('video count vs Subscribers')
plt.show()

# Correlation analysis
correlation = df['video count'].corr(df['subscribers'])
print(f'Correlation between video count and Subscribers: {correlation}')
```

```python
category_views=df.groupby('category')['video views'].mean().sort_values(ascending=False)   #using groupby funtion to determine avg views for each category
```

```python
category_views=category_views.reset_index()
```

```python
category_views.rename(columns={'video views':'avg_views'},inplace=True)
```

```python
category_views
```

```python
category_views.plot(kind='bar',x='category',y='avg_views')  #shows is the most watched by the viewers
plt.title('Average Video Views by Category')
plt.ylabel('avg_views')
plt.show()
```

```python
df['upload_frequency'] = df['video count']/df['No_of_years']          #correlation bewteen upload frequency and the video views
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='upload_frequency', y='video views',palette=)
plt.title('Upload Frequency vs Average Video Views')
plt.xlabel('Upload Frequency')
plt.ylabel('Average Video Views')
plt.show()

correlation = df['upload_frequency'].corr(df['video views'])
print(f'Correlation between upload frequency and video views: {correlation}')
```

```python
correlation = df['subscribers'].corr(df['video views']).mean()
print(f'Correlation between subscribers and average video views: {correlation}')
```

```python
category_counts = df.groupby('category')['Youtuber'].count().sort_values(ascending=False) #Number of Different Category Types Each Channel Creates Content For
print(category_counts)                                                                                         #most of the channels are in entertainment category
```

```python
df['period'] = pd.cut(df['started'], bins=[0, 2010, 2015, 2024], labels=['Before 2010', 'Between 2010-2015', 'After 2015'])   #video count across different periods
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='period', y='video count',palette='viridis')
plt.title('Video Count Across Different Periods')
plt.xlabel('Period')
plt.xticks(rotation=45)
plt.ylabel('Video Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='upload_frequency', y='category', data=df, orient='h')   #video upload frequency across each category
plt.title('Video Upload Frequency by Category')
plt.xlabel('Video Count')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='category', y='growth_rate')    #growth rate of subscribers across each category
plt.xticks(rotation=90)
plt.title('Subscriber Growth Rate Across Categories')
plt.xlabel('Category')
plt.ylabel('Subscriber Growth Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='rank', y='video count')   #rank vs video count graph with correlation
plt.title('Rank vs Video Count')
plt.xlabel('Rank')
plt.ylabel('Video Count')
plt.show()

correlation = df['rank'].corr(df['video count'])
print(f'Correlation between rank and video count: {correlation}')
```

```python
early_channels = df[df['started'] < 2010]
recent_channels = df[df['started'] >= 2010]
early_avg_subs = early_channels['subscribers'].mean()
recent_avg_subs = recent_channels['subscribers'].mean()
early_avg_views = early_channels['video views'].mean()
recent_avg_views = recent_channels['video views'].mean()
print(f'Average Subscribers (Early): {early_avg_subs}, (Recent): {recent_avg_subs}')
print(f'Average Video Views (Early): {early_avg_views}, (Recent): {recent_avg_views}')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rank', y='video views', data=df)   #relationship between rank and video views with correlation
plt.title('Relationship between Rank and Average Video Views')
plt.xlabel('Rank')
plt.ylabel('Average Video Views')
plt.show()

correlation = df['rank'].corr(df['video views'])
print(f'Correlation between rank and video views: {correlation}')
```
